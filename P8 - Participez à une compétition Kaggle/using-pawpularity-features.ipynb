```python
import pandas as pd
from tqdm import tqdm
chunksize = 100
X_path = "/kaggle/input/pawpularity-features-targets/X_features.csv"
dfs = []
TextFileReader = pd.read_csv(X_path, header=None, chunksize=chunksize)
for chunk_df in tqdm(TextFileReader):
    dfs.append(chunk_df)
```

```python
df = pd.concat(dfs, sort=False)
```

```python
df
```

```python
y_path = "/kaggle/input/pawpularity-features-targets/y_targets.csv"
df_target = pd.read_csv(y_path, header=None)
df_target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, df_target, test_size=0.2, random_state=42)
```

```python
# free memory
import os, psutil, gc
def usage():
    process = psutil.Process(os.getpid())
    print(process.memory_info()[0] / float(2 ** 20))
usage()
del df
del df_target
del dfs
gc.collect()
usage()
```
