On examine les colonnes de ces 2 fichiers pour estimer lequel est le plus intéressant ou s'ils sont complémentaires

```python
print(data2015.columns)
print(data2016.columns)
```

```python
def showDiffsColums():
    nbColsManquantesDans2016 = 0
    nbColsManquantesDans2015 = 0
    for col in data2015.columns:
        if col not in data2016.columns:
            print("- colonne", col, "présente dans le fichier 2015 mais pas dans le 2016")
            print(data2015[col].head(), "\n")
            nbColsManquantesDans2016 += 1
    print('-------------')
    for col in data2016.columns:
        if col not in data2015.columns:
            print("- colonne", col, "présente dans le fichier 2016 mais pas dans le 2015")
            print(data2016[col].head(), "\n")
            nbColsManquantesDans2015 += 1
    print('-------------')
    print("Nombre de colonnes manquantes dans le fichier 2015 :", nbColsManquantesDans2015)
    print("Nombre de colonnes manquantes dans le fichier 2016 :", nbColsManquantesDans2016)
showDiffsColums()
```

Parmi les colonnes manquantes, on constate que la colonne "Location" du fichier 2015 contient des informations sérialisées en JSON : on va les désérialiser et rajouter les données sous la forme de colonnes portant le même nom que dans le fichier 2016.

```python
print(data2015['Location'][0])
```

```python
import json
latitude = []
longitude = []
address = []
city = []
state = []
zipcode = []
for location in data2015['Location']:
    loc = json.loads(location.replace('"', '\\"').replace("'", '"'))
    human_address = json.loads(loc['human_address'])
    latitude.append(loc['latitude'])
    longitude.append(loc['longitude'])
    address.append(human_address['address'])
    city.append(human_address['city'])
    state.append(human_address['state'])
    zipcode.append(human_address['zip'])
data2015['Latitude'] = latitude
data2015['Longitude'] = longitude
data2015['Address'] = address
data2015['City'] = city
data2015['State'] = state
data2015['ZipCode'] = zipcode
data2015 = data2015.drop(['Location'], axis=1)
print(data2015.shape)
showDiffsColums()
```

Nous sommes passé de 47 à 52 colonnes (6 ajouts, 1 suppression) et de 9 à 3 colonnes manquantes dans le fichier 2015.

On va s'intéresser maintenant au taux de remplissage des colonnes pour chaque jeu de données :

```python
import matplotlib.pyplot as plt
datas = { '2015': data2015, '2016': data2016 }
for year in datas.keys():
    data = datas[year]
    percent_missing = data.isnull().sum() * 100 / len(data)
    missing_values = pd.DataFrame({'pourcentage_manquant': percent_missing})
    missing_values.sort_values('pourcentage_manquant', inplace=True)
    ax1 = missing_values["pourcentage_manquant"].plot(kind='bar')
    ax1.xaxis.set_visible(False)
    plt.title("données de " + year)
    plt.ylabel("pourcentage de valeurs manquantes")
    plt.show()
    print(missing_values[missing_values["pourcentage_manquant"] > 10])
```

Je cherche maintenant à mieux connaitre les différentes valeurs saisies dans les colonnes non numériques de notre jeu de données :

```python
import numpy as np
datas = { '2015': data2015, '2016': data2016 }
cols = ['BuildingType', 'PrimaryPropertyType', 'Neighborhood', 'YearBuilt', 'NumberofBuildings', 'NumberofFloors', 'LargestPropertyUseType', 'ComplianceStatus', 'ENERGYSTARScore']
cols_index_sort = ['YearBuilt', 'NumberofBuildings', 'NumberofFloors', 'ENERGYSTARScore'] # les colonnes à trier en fonction de l'index

def show_distribution_col(col):
    plt.rcParams['figure.figsize']=(10,6)
    for year in datas.keys():
        data = datas[year]
        uniques = data[~data[col].isna()][col].unique()
        uniques.sort()
        print(col, year, ":\n", uniques, "\n")
        if col in cols_index_sort:
            groups_count = data[~data[col].isna()][col].value_counts(normalize=True, sort=False).sort_index(ascending=True)
        else:
            groups_count = data[~data[col].isna()][col].value_counts(normalize=True, sort=True)
        if col != 'LargestPropertyUseType': # pour une question de lisibilité
            plt.subplot(1, 2, int(year) - 2014) # pour avoir 2015 en 1 et 2016 en 2
        groups_count.plot(kind='bar')
        if col in ['YearBuilt', 'NumberofFloors', 'ENERGYSTARScore']:
            plt.locator_params(axis='x', nbins=20) # pour une question de lisibilité
        plt.xlabel(col + " " + year)
        if col == 'LargestPropertyUseType' and year == '2015': # pour une question de lisibilité
            plt.show()
    plt.show()
    plt.rcdefaults()

for col in cols:
    show_distribution_col(col)
```

Les graphiques précédents nous montrent des données de casses différentes dans la colonne 'Neighborhood' des données de 2016. On a également une valeur 'DELRIDGE NEIGHBORHOODS' au lieu de juste 'DELRIDGE' : on uniformise.

```python
data2016['Neighborhood'] = data2016['Neighborhood'].str.upper()
delridge = data2016[data2016['Neighborhood'] == 'DELRIDGE NEIGHBORHOODS']['Neighborhood']
data2016.loc[delridge.index, ['Neighborhood']] = 'DELRIDGE'
show_distribution_col('Neighborhood')
```

Je cherche à comprendre la colonne "ComplianceStatus" - la description nous dit "Whether a property has met energy benchmarking requirements for the current reporting year." mais ce n'est pas très clair pour moi.

```python
datas = { '2015': data2015, '2016': data2016 }
for year in datas.keys():
    print("\n---", year, "---")
    data = datas[year]
    isCompliant = data[data['ComplianceStatus'] == 'Compliant']
    isNotCompliant = data[data['ComplianceStatus'] != 'Compliant']
    print("Valeurs d'ENERGYSTARScore uniques là où on a 'Compliant' :\n", isCompliant['ENERGYSTARScore'].unique())
    print("Valeurs d'ENERGYSTARScore uniques là où on n'a pas 'Compliant' :", isNotCompliant['ENERGYSTARScore'].unique())
    print("Nombre de valeurs d'ENERGYSTARScore renseignées là où on a 'Compliant' :", len(isCompliant[isCompliant['ENERGYSTARScore'].notna()]))
    print("Nombre de valeurs d'ENERGYSTARScore non renseignées là où on a 'Compliant' :", len(isCompliant[isCompliant['ENERGYSTARScore'].isna()]))
    #if year == '2016':
    print("Nombre de valeurs d'ENERGYSTARScore renseignées là où on a pas 'Compliant' :", len(isNotCompliant[isNotCompliant['ENERGYSTARScore'].notna()]))
    print("Nombre de valeurs d'ENERGYSTARScore non renseignées là où on a pas 'Compliant' :", len(isNotCompliant[isNotCompliant['ENERGYSTARScore'].isna()]))
```

Il semble donc que la notion de "Compliant" permet de savoir que l'ENERGYSTARScore est calculable, mais pas forcément calculé :
- en 2015, 2560 ENERGYSTARScore ont été calculés sur les 3337 bâtiments conformes, soit environ 77%.
- en 2016, 2399 ENERGYSTARScore ont été calculés sur les 3211 bâtiments conformes, soit environ 75%.

Etrangement, on a moins de bâtiments conformes en 2016 et encore moins de scores calculés, mais on a des valeurs sur des bâtiments non conformes... peut-être issues des calculs de l'année précédente ?
